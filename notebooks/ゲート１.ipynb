# **量子ゲート講座 第一回（エントリーコース）おさらい**

## **1. 量子ビット**

量子コンピュータでは情報の単位として**量子ビット(qubit: キュービット)**を考えました。$\left|0\right>$, $\left|1\right>$ という状態が基本でした。

\
量子ビットは古典ビットとは異なり $\left|0\right>$, $\left|1\right>$ 以外にもさまざまな状態をとることができました。

量子コンピュータワークショップ2023_量子ビット.png

### **量子の重ね合わせ**

量子ビットはあらゆる角度に動かすことができ、それぞれ異なる状態をとることができます。

量子コンピュータワークショップ2023_重ね合わせ.png

### **量子のもつれ（エンタングルメント）: 今日は使わない**

2量子ビットの場合は通常はばらばらに状態がとれます。

## **2. 量子コンピュータの計算**

量子ビットに何かしらの操作を加えることで量子コンピュータは計算を実行します。前回はX、H、CXという操作 (ゲート) を考えました。

### **Xゲート**

$|0 \rangle$を$|1 \rangle$に、$|1 \rangle$を$|0 \rangle$にするのでした。

x.png

### **Hゲート**

回転して横向きになった矢印の状態を作るのでした。結果、$|0 \rangle$と$|1 \rangle$が半分ずつ重ね合わさった状態になりました。

h.png

### **Hゲート (ご質問をいただいたケース)**

Hゲートの手前にXゲートを入れても結果に変化がないように見えます。実は異なるのですが、エントリーの範囲では扱わない知識が必要になります。アドバンスコースにて扱われると思いますのでお待ちください。

h_2.png

### **CXゲート**

制御ビットが$| 1 \rangle$の時、ターゲットビットにXゲートが適用されるのでした。

cx.png
